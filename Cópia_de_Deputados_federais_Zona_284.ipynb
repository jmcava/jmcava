```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Caminho do arquivo
arquivo = '/content/votacao_candidato.csv'

# Ler o arquivo CSV com codificação 'latin1'
dados = pd.read_csv(arquivo, sep=';', encoding='latin1')

# Exibir as primeiras linhas para verificar o conteúdo
print(dados.head())

# Filtragem para candidatos de um cargo específico (Deputado Federal)
deputados_federais = dados[dados['Cargo'] == 'Deputado Federal']

# Filtragem de candidatos com votos acima de 100 e até 11.000
deputados_federais_filtrados = deputados_federais[(deputados_federais['Votos nominais'] > 1) &
                                                 (deputados_federais['Votos nominais'] <= 11000)]

# Ordenar os candidatos por votos nominais
deputados_federais_filtrados = deputados_federais_filtrados.sort_values(by='Votos nominais', ascending=False)

# Dividir os dados em blocos de 25 candidatos
num_blocos = int(np.ceil(len(deputados_federais_filtrados) / 25))

# Gerar gráfico para cada bloco
for i in range(num_blocos):
    # Determinar o índice de início e fim para o bloco
    start_idx = i * 25
    end_idx = (i + 1) * 25

    # Filtrar o bloco atual de candidatos
    bloco_atual = deputados_federais_filtrados.iloc[start_idx:end_idx]

    # Criar o gráfico de barras horizontais para o bloco
    plt.figure(figsize=(10, 8))
    bars = plt.barh(bloco_atual['Nome candidato'], bloco_atual['Votos nominais'], color='skyblue')

    # Adicionar rótulos de valores nas barras
    for bar in bars:
        plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{int(bar.get_width())}',
                 va='center', ha='left', color='black', fontsize=10)

    # Títulos e labels
    plt.xlabel('Votos Nominais')
    plt.ylabel('Candidato')
    plt.title(f'Bloco {i + 1}: Candidatos Mais Votados - Deputado Federal em São Bernardo do Campo - Zona 284')
    plt.gca().invert_yaxis()  # Inverter a ordem para o maior valor no topo

    # Exibir o gráfico
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages

# Caminho do arquivo
arquivo = '/content/votacao_candidato.csv'

# Ler o arquivo CSV com codificação 'latin1'
dados = pd.read_csv(arquivo, sep=';', encoding='latin1')

# Filtragem para candidatos de um cargo específico (Deputado Federal)
deputados_federais = dados[dados['Cargo'] == 'Deputado Federal']

# Filtragem de candidatos com votos acima de 100 e até 11.000
deputados_federais_filtrados = deputados_federais[(deputados_federais['Votos nominais'] > 1) &
                                                 (deputados_federais['Votos nominais'] <= 11000)]

# Ordenar os candidatos por votos nominais
deputados_federais_filtrados = deputados_federais_filtrados.sort_values(by='Votos nominais', ascending=False)

# Dividir os dados em blocos de 25 candidatos
num_blocos = int(np.ceil(len(deputados_federais_filtrados) / 25))

# Criar um PDF para salvar os gráficos
with PdfPages('/content/grafico_votos_deputados_federais.pdf') as pdf:
    # Gerar gráfico para cada bloco
    for i in range(num_blocos):
        # Determinar o índice de início e fim para o bloco
        start_idx = i * 25
        end_idx = (i + 1) * 25

        # Filtrar o bloco atual de candidatos
        bloco_atual = deputados_federais_filtrados.iloc[start_idx:end_idx]

        # Criar o gráfico de barras horizontais para o bloco
        plt.figure(figsize=(10, 8))
        bars = plt.barh(bloco_atual['Nome candidato'], bloco_atual['Votos nominais'], color='skyblue')

        # Adicionar rótulos de valores nas barras
        for bar in bars:
            plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{int(bar.get_width())}',
                     va='center', ha='left', color='black', fontsize=10)

        # Títulos e labels
        plt.xlabel('Votos Nominais')
        plt.ylabel('Candidato')
        plt.title(f'Bloco {i + 1}: Candidatos Mais Votados - Deputado Federal em São Bernardo do Campo - Zona 284')
        plt.gca().invert_yaxis()  # Inverter a ordem para o maior valor no topo

        # Salvar o gráfico no PDF
        pdf.savefig()  # Salvar a figura na página do PDF
        plt.close()  # Fechar a figura para o próximo gráfico

print("PDF gerado com sucesso!")
```
