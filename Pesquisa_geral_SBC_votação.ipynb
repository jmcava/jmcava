```python
import matplotlib.pyplot as plt
import pandas as pd

# Dados de exemplo
data = {
    "Zona": [174, 283, 284, 296, 409],
    "Quantidade de aptos": [111279, 107327, 106473, 103344, 107129],
    "Quantidade de votos totais": [83855, 79110, 75577, 77440, 80863],
    "Quantidade de seções": [402, 316, 314, 297, 320],
    "Quantidade de votos nominais válidos": [74316, 69435, 65639, 68518, 70801],
    "Quantidade de abstenções": [27424, 28217, 30896, 25904, 26266],
    "Quantidade de votos brancos": [3520, 3355, 3314, 3103, 3671],
    "Quantidade de votos nulos": [6019, 6320, 6624, 5819, 6391],
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Total de eleitores
total_eleitores = 643027

# Definindo as categorias
categories = [
    'Quantidade de aptos',
    'Quantidade de votos totais',
    'Quantidade de seções',
    'Quantidade de votos nominais válidos',
    'Quantidade de abstenções',
    'Quantidade de votos brancos',
    'Quantidade de votos nulos'
]

# Calculando a soma geral para cada índice
totals = df[categories].sum()

# Calculando a porcentagem em relação ao total de eleitores
percentages = (totals / total_eleitores) * 100

# Configurando o gráfico
plt.figure(figsize=(10, 6))

# Definindo cores para cada barra
colors = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99', '#FFD700', '#C0C0C0', '#FF4500']

# Criando o gráfico de barras para porcentagens
bars = plt.bar(percentages.index, percentages.values, color=colors)

# Adicionando valores na ponta das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{yval:.2f}%', ha='center', va='bottom')

# Adicionando título e legendas
plt.title('Porcentagem das Quantidades em Relação ao Total de Eleitores - SBCampo 2024 = 643.023 eleitores')
plt.xlabel('Índices')
plt.ylabel('Porcentagem (%)')
plt.xticks(rotation=45, ha='right')

# Exibindo o gráfico
plt.tight_layout()
plt.grid(axis='y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Novos dados fornecidos
data = {
    "Zona": [174, 283, 284, 296, 409],
    "Quantidade de votos brancos": [3367, 3446, 3169, 3251, 3325],
    "Quantidade de votos válidos": [65125, 61363, 59081, 60434, 62789],
    "Quantidade total de seções": [402, 316, 314, 298, 320],
    "Quantidade de votos nulos": [7742, 8030, 7987, 7456, 7031],
    "Quantidade de votos totais": [76234, 72839, 70237, 71141, 73145],
    "Quantidade de aptos": [111279, 107328, 106473, 103344, 107129],
    "Quantidade de abstenções": [35045, 34489, 36236, 32203, 33984],
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Configurando o gráfico
plt.figure(figsize=(14, 8))

# Definindo a largura das barras
bar_width = 0.1

# Definindo a posição das barras
positions = range(len(df))

# Ordem das barras
categories = [
    'Quantidade de aptos',
    'Quantidade de votos totais',
    'Quantidade total de seções',
    'Quantidade de votos válidos',
    'Quantidade de abstenções',
    'Quantidade de votos brancos',
    'Quantidade de votos nulos',
]

# Criando as barras para cada categoria
for i, category in enumerate(categories):
    plt.bar([p + i * bar_width for p in positions], df[category], width=bar_width, label=category)

# Adicionando valores na ponta das barras
for i, category in enumerate(categories):
    for bar in plt.bar([p + i * bar_width for p in positions], df[category], width=bar_width):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')

# Definindo o eixo x com os nomes das zonas
plt.xticks([p + bar_width * (len(categories) - 1) / 2 for p in positions], df['Zona'])

# Adicionando título e legendas
plt.title('Eleição Municipal 2024 para Prefeito - SBCampo 2º Turno Zonas 174, 283, 284, 296, 409')
plt.xlabel('Zonas Eleitorais')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(axis='y')

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Novos dados fornecidos
data = {
    "Zona": [174, 283, 284, 296, 409],
    "Quantidade de votos brancos": [3367, 3446, 3169, 3251, 3325],
    "Quantidade de votos válidos": [65125, 61363, 59081, 60434, 62789],
    "Quantidade total de seções": [402, 316, 314, 298, 320],
    "Quantidade de votos nulos": [7742, 8030, 7987, 7456, 7031],
    "Quantidade de votos totais": [76234, 72839, 70237, 71141, 73145],
    "Quantidade de aptos": [111279, 107328, 106473, 103344, 107129],
    "Quantidade de abstenções": [35045, 34489, 36236, 32203, 33984],
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Configurando o gráfico
plt.figure(figsize=(14, 8))

# Definindo a largura das barras
bar_width = 0.1

# Definindo a posição das barras
positions = range(len(df))

# Ordem das barras
categories = [
    'Quantidade de aptos',
    'Quantidade de votos totais',
    'Quantidade total de seções',
    'Quantidade de votos válidos',
    'Quantidade de abstenções',
    'Quantidade de votos brancos',
    'Quantidade de votos nulos',
]

# Definindo cores para cada categoria
colors = [
    'blue',    # Quantidade de aptos
    'green',   # Quantidade de votos totais
    'orange',  # Quantidade total de seções
    'red',     # Quantidade de votos válidos
    'purple',  # Quantidade de abstenções
    'cyan',    # Quantidade de votos brancos
    'magenta'  # Quantidade de votos nulos
]

# Criando as barras para cada categoria
for i, category in enumerate(categories):
    plt.bar([p + i * bar_width for p in positions], df[category], width=bar_width, label=category, color=colors[i])

# Adicionando valores na ponta das barras
for i, category in enumerate(categories):
    for bar in plt.bar([p + i * bar_width for p in positions], df[category], width=bar_width, color=colors[i]):
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, yval, ha='center', va='bottom')

# Definindo o eixo x com os nomes das zonas
plt.xticks([p + bar_width * (len(categories) - 1) / 2 for p in positions], df['Zona'])

# Adicionando título e legendas
plt.title('Eleição Municipal 2024 para Prefeito - SBCampo 2º Turno Zonas 174, 283, 284, 296, 409')
plt.xlabel('Zonas Eleitorais')
plt.ylabel('Quantidade')
plt.legend()
plt.grid(axis='y')

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
