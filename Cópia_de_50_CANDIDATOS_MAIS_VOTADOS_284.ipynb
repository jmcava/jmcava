```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que você tenha os dados em um DataFrame chamado df
df = pd.read_csv('/content/votacao_local_votacao-zona_2024_sp_284.csv')

# 1. Calcular Taxa de Comparecimento e Abstenção
df['taxa_comparecimento'] = (df['qt_comparecimento'] / df['qt_aptos']) * 100
df['taxa_abstencao'] = (df['qt_abstencoes'] / df['qt_aptos']) * 100

# Visualizar taxas por zona eleitoral
plt.figure(figsize=(10, 6))
plt.bar(df['nr_zona'], df['taxa_comparecimento'], label='Comparecimento', alpha=0.7)
plt.bar(df['nr_zona'], df['taxa_abstencao'], label='Abstenção', alpha=0.7)
plt.xlabel('Zona Eleitoral')
plt.ylabel('Taxa (%)')
plt.title('Taxa de Comparecimento e Abstenção por Zona Eleitoral')
plt.legend()
plt.show()

# 2. Analisar votos válidos por candidato
plt.figure(figsize=(10, 6))
for candidato in df['nm_votavel'].unique():
    votos_candidato = df[df['nm_votavel'] == candidato]
    plt.plot(votos_candidato['nr_zona'], votos_candidato['qt_votos_nominais'], label=candidato)

plt.xlabel('Zona Eleitoral')
plt.ylabel('Votos Válidos')
plt.title('Votos Válidos por Candidato por Zona Eleitoral')
plt.legend()
plt.show()

# 3. Analisar o total de votos por zona
plt.figure(figsize=(10, 6))
plt.bar(df['nr_zona'], df['qt_votos'])
plt.xlabel('Zona Eleitoral')
plt.ylabel('Total de Votos')
plt.title('Total de Votos por Zona Eleitoral')
plt.show()
```

```python
import pandas as pd

# Carregar o CSV com a codificação 'latin1' e o separador correto (como ';' ou ',')
df = pd.read_csv('/content/votacao_local_votacao-zona_2024_sp_284.csv', encoding='latin1', sep=';')

# Verificar as primeiras linhas do DataFrame para entender a estrutura
df.head()
```

```python
# Selecionando as colunas necessárias
columns_to_use = ['ds_eleicao', 'nr_zona', 'nr_votavel', 'nm_votavel', 'qt_aptos', 'qt_comparecimento', 'qt_abstencoes', 'qt_votos_nominais', 'qt_votos']

df_selected = df[columns_to_use]

# Verificando as primeiras linhas do DataFrame selecionado
df_selected.head()
```

```python
# Verificando valores ausentes nas colunas selecionadas
df_selected.isnull().sum()
```

```python
# Preenchendo valores ausentes com 0 (ou outro valor se necessário)
df_selected = df_selected.fillna(0)

# Verificando novamente para garantir que não há valores nulos
df_selected.isnull().sum()
```

```python
# Calcular Taxa de Comparecimento
df_selected['taxa_comparecimento'] = df_selected['qt_comparecimento'] / df_selected['qt_aptos'] * 100

# Calcular Taxa de Abstenção
df_selected['taxa_abstencao'] = df_selected['qt_abstencoes'] / df_selected['qt_aptos'] * 100

# Exibir as primeiras linhas com as novas colunas
df_selected.head()
```

```python
# Analisando a média das taxas
media_taxa_comparecimento = df_selected['taxa_comparecimento'].mean()
media_taxa_abstencao = df_selected['taxa_abstencao'].mean()

print(f'Média da Taxa de Comparecimento: {media_taxa_comparecimento:.2f}%')
print(f'Média da Taxa de Abstenção: {media_taxa_abstencao:.2f}%')
```

```python
# Salvando o DataFrame com as novas colunas em um novo arquivo CSV
df_selected.to_csv('/content/votacao_processada.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Passo 1: Dividir os 50 primeiros candidatos em dois grupos
top_25 = df_candidatos_top_50.head(25)
top_26_50 = df_candidatos_top_50.tail(25)

# Passo 2: Criar o gráfico para os 25 primeiros candidatos
plt.figure(figsize=(12, 8))
bars_25 = plt.barh(top_25['nm_votavel'], top_25['qt_votos'], color='skyblue')
plt.xlabel('Total de Votos')
plt.ylabel('Candidato')
plt.title('Candidatos 1 a 25 Mais Votados (Zona 284)')
plt.gca().invert_yaxis()  # Para inverter a ordem e exibir o mais votado no topo

# Adicionar os valores nas barras
for bar in bars_25:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{int(bar.get_width())}',
             va='center', ha='left', color='black', fontweight='bold')

plt.show()

# Passo 3: Criar o gráfico para os candidatos de 26 a 50
plt.figure(figsize=(12, 8))
bars_26_50 = plt.barh(top_26_50['nm_votavel'], top_26_50['qt_votos'], color='lightcoral')
plt.xlabel('Total de Votos')
plt.ylabel('Candidato')
plt.title('Candidatos 26 a 50 Mais Votados (Zona 284)')
plt.gca().invert_yaxis()  # Para inverter a ordem e exibir o mais votado no topo

# Adicionar os valores nas barras
for bar in bars_26_50:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height() / 2,
             f'{int(bar.get_width())}',
             va='center', ha='left', color='black', fontweight='bold')

plt.show()
```
