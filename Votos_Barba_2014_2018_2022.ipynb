```python
import pandas as pd

# Caminho do arquivo CSV
file_path = "/content/votacao_candidato.csv"

# Carregar o arquivo CSV em um DataFrame
df = pd.read_csv(file_path, sep=';', encoding='latin1')

# Exibir as primeiras linhas para conferência
print("Visualização das primeiras linhas:")
print(df.head())

# Filtrar os dados apenas para o estado de São Paulo, cargo de Deputado Estadual, e o candidato Teonilio Monteiro da Costa
df_filtered = df[
    (df['UF'] == 'SP') &
    (df['Cargo'] == 'Deputado Estadual') &
    (df['Nome candidato'] == 'TEONILIO MONTEIRO DA COSTA')
]

# Selecionar as colunas principais para análise
df_filtered = df_filtered[['Ano de eleição', 'Cargo', 'UF', 'Nome candidato', 'Votos nominais']]

# Exibir o DataFrame filtrado para conferência
print("\nDados filtrados:")
print(df_filtered)

# Agrupar por ano de eleição e calcular o total de votos
votos_agrupados = df_filtered.groupby('Ano de eleição')['Votos nominais'].sum().reset_index()

# Renomear as colunas para facilitar a leitura
votos_agrupados.columns = ['Ano de Eleição', 'Total de Votos']

# Exibir o total de votos por ano
print("\nTotal de votos por ano de eleição:")
print(votos_agrupados)

# Opcional: Salvar os resultados em um novo arquivo CSV
output_path = "/content/votos_teonilio_monteiro_da_costa.csv"
votos_agrupados.to_csv(output_path, index=False, encoding='utf-8-sig')
print(f"\nArquivo com resultados salvo em: {output_path}")
```

```python
import pandas as pd

# Caminho do arquivo
file_path_2022 = "/content/votacao_candidato_2022.csv"

# Carregar o arquivo
df_2022 = pd.read_csv(file_path_2022, encoding='latin1', sep=';')

# Normalizar os nomes das colunas (remover espaços e ajustar textos)
df_2022.columns = [col.strip() for col in df_2022.columns]

# Garantir que os textos nas colunas relevantes estejam normalizados
df_2022['UF'] = df_2022['UF'].str.strip().str.upper()
df_2022['Cargo'] = df_2022['Cargo'].str.strip().str.upper()
df_2022['Nome candidato'] = df_2022['Nome candidato'].str.strip().str.upper()

# Valores de filtragem também normalizados
uf = "SP"
cargo = "DEPUTADO ESTADUAL"
nome_candidato = "TEONÍLIO MONTEIRO DA COSTA"

# Filtrar os dados
df_filtered_2022 = df_2022[
    (df_2022['UF'] == uf) &
    (df_2022['Cargo'] == cargo) &
    (df_2022['Nome candidato'] == nome_candidato)
]

# Selecionar apenas as colunas principais
df_filtered_2022 = df_filtered_2022[['Ano de eleição', 'Cargo', 'UF', 'Nome candidato', 'Votos nominais']]

# Exibir os dados filtrados
print("\nDados filtrados para 2022:")
print(df_filtered_2022)

# Verificar se há dados antes de agrupar
if not df_filtered_2022.empty:
    # Agrupar por ano de eleição e calcular o total de votos
    votos_agrupados_2022 = df_filtered_2022.groupby('Ano de eleição')['Votos nominais'].sum().reset_index()

    # Renomear as colunas para facilitar a leitura
    votos_agrupados_2022.columns = ['Ano de Eleição', 'Total de Votos']

    # Exibir o total de votos por ano de eleição
    print("\nTotal de votos por ano de eleição (2022):")
    print(votos_agrupados_2022)

    # Salvar os resultados
    output_path_2022 = "/content/resultados_teonilio_2022.csv"
    votos_agrupados_2022.to_csv(output_path_2022, index=False, encoding='utf-8-sig')
    print(f"\nArquivo com resultados de 2022 salvo em: {output_path_2022}")
else:
    print("\nNenhum dado encontrado após a filtragem. Verifique os valores de filtragem ou os dados do arquivo.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados das votações de 2014, 2018 e 2022
dados_votacao = {
    "Ano": [2014, 2018, 2022],
    "Total de Votos": [96156, 91394, 108071]
}

# Criar um DataFrame a partir dos dados
df_votacao = pd.DataFrame(dados_votacao)

# Configurar o gráfico
plt.figure(figsize=(8, 6))
plt.bar(df_votacao["Ano"], df_votacao["Total de Votos"], color=['blue', 'green', 'red'])

# Adicionar título e rótulos
plt.title("Votação de Teonílio Monteiro da Costa (2014, 2018, 2022)", fontsize=14)
plt.xlabel("Ano de Eleição", fontsize=12)
plt.ylabel("Total de Votos", fontsize=12)

# Adicionar valores em cima das barras
for index, value in enumerate(df_votacao["Total de Votos"]):
    plt.text(df_votacao["Ano"][index], value + 2000, f"{value:,}".replace(",", "."), ha='center', fontsize=10)

# Exibir o gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados das votações de 2014, 2018 e 2022
dados_votacao = {
    "Ano": [2014, 2018, 2022],
    "Total de Votos": [96156, 91394, 108071]
}

# Criar um DataFrame a partir dos dados
df_votacao = pd.DataFrame(dados_votacao)

# Calcular as diferenças de votos
df_votacao['Diferença de Votos'] = df_votacao['Total de Votos'].diff().fillna(0)

# Mostrar as diferenças
print("Diferenças nos votos por ano:")
print(df_votacao[['Ano', 'Diferença de Votos']])

# Configurar o gráfico para mostrar as diferenças
plt.figure(figsize=(8, 6))
plt.bar(df_votacao["Ano"][1:], df_votacao["Diferença de Votos"][1:], color=['orange', 'purple'])

# Adicionar título e rótulos
plt.title("Diferença de Votos para Teonílio Monteiro da Costa (2014, 2018, 2022)", fontsize=14)
plt.xlabel("Ano de Eleição", fontsize=12)
plt.ylabel("Diferença de Votos", fontsize=12)

# Adicionar valores em cima das barras
for index, value in enumerate(df_votacao["Diferença de Votos"][1:]):
    plt.text(df_votacao["Ano"][index + 1], value + 500, f"{value:,.0f}".replace(",", "."), ha='center', fontsize=10)

# Exibir o gráfico
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Relatório sobre as diferenças de votos
relatorio = """
Relatório de Diferença de Votos para Teonílio Monteiro da Costa:

1. Em 2018, comparado com 2014, houve uma diminuição de 4.762 votos. Isso representa uma redução de aproximadamente 4,95% em relação ao número de votos obtidos em 2014.

2. No entanto, em 2022, comparado com 2018, houve um aumento significativo de 16.677 votos, o que equivale a um crescimento de cerca de 18,24% em relação a 2018.

Esses resultados indicam que, após uma queda nas eleições de 2018, houve um crescimento robusto na votação de 2022, o que pode ser um reflexo de estratégias de campanha, mudanças no cenário político, ou maior reconhecimento do candidato ao longo do tempo.

Conclusão: A votação de Teonílio Monteiro da Costa mostra uma recuperação em 2022 após a queda em 2018. A análise das tendências eleitorais é essencial para ajustar as futuras estratégias de campanha e maximizar a capilaridade do candidato.
"""

print(relatorio)
```
