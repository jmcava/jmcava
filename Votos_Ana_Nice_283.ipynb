```python
import pandas as pd
import matplotlib.pyplot as plt

# --- Configuração ---
ANO_ELEICAO = 2020  # Altere se necessário
NOME_CANDIDATA = "Ana Nice Martins Carvalho"
ZONA_ELEITORAL = 283
TOTAL_VOTOS_ANA_NICE = 903  # Total de votos da Ana Nice na zona 283

# --- Lista de escolas da zona 283 (complete com todas as escolas) ---
ESCOLAS_ZONA_283 = [
    "EE JORNALISTA VLADIMIR HERZOG",
    "EE, ENGENHEIRO FRANCISCO PRESTES MAIA",
    "EE, MIZUHO",
    "EE, PROF JOAO BATISTA BERNARDES",
    "EE, PROFA, MARIA LUÍZA FERRARI CICERO",
    # ... adicione as demais escolas ...
    "IMS, INSTITUTO METODISTA DE ENSINO SUPERIOR"
]

# --- Criação do DataFrame inicial (sem os votos por escola ainda) ---
df_votos = pd.DataFrame({
    'nm_local_votacao': ESCOLAS_ZONA_283,
    'qt_votos': [0] * len(ESCOLAS_ZONA_283),  # Inicializa com zero - será preenchido depois
    'ano_eleicao': [ANO_ELEICAO] * len(ESCOLAS_ZONA_283),
    'zona_eleitoral': [ZONA_ELEITORAL] * len(ESCOLAS_ZONA_283),
    'nome_candidato': [NOME_CANDIDATA] * len(ESCOLAS_ZONA_283)
})


# --- Função para atualizar o DataFrame e plotar o gráfico ---
def atualizar_e_plotar(df_votos, total_votos, arquivo_votos="votos_zona_283.csv"):
    try:
        df_votos_local = pd.read_csv(arquivo_votos, encoding='latin1', sep=';') # Ajuste o separador se necessário
    except FileNotFoundError:
        print(f"Erro: Arquivo '{arquivo_votos}' não encontrado.")
        return
    except Exception as e:
        print(f"Erro ao ler o arquivo: {e}")
        return

    # Mesclar com o df_votos original para manter todas as escolas
    df_votos = pd.merge(df_votos, df_votos_local[['nm_local_votacao', 'qt_votos']], on='nm_local_votacao', how='left')

    # Preencher valores NaN com 0 (caso alguma escola não tenha votos no arquivo)
    df_votos['qt_votos'] = df_votos['qt_votos'].fillna(0)

    # Ordenar por quantidade de votos
    df_ordenado = df_votos.sort_values('qt_votos', ascending=True)


    # Plotar o gráfico
    plt.figure(figsize=(12, 8))
    plt.barh(df_ordenado['nm_local_votacao'], df_ordenado['qt_votos'], color='skyblue')
    for index, value in enumerate(df_ordenado['qt_votos']):
        plt.text(value, index, str(value), va='center', ha='left', color='black', fontsize=10)

    plt.xlabel('Quantidade de Votos', fontsize=12)
    plt.ylabel('Local de Votação', fontsize=12)
    plt.title(f'Votos de {NOME_CANDIDATA} na Zona {ZONA_ELEITORAL} ({ANO_ELEICAO}) - Total: {total_votos}', fontsize=14)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.tight_layout()
    plt.show()


# --- Chame a função quando tiver o arquivo com os votos ---
atualizar_e_plotar(df_votos, TOTAL_VOTOS_ANA_NICE)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np

# --- Configuração ---
ANO_ELEICAO = 2024
NOME_CANDIDATA = "ANA NICE MARTINS DE CARVALHO"
ZONA_ELEITORAL = 283
TOTAL_VOTOS_ANA_NICE = 903  # Movido para o cabeçalho

# --- Carregamento dos dados ---
try:
    df_votos_secao = pd.read_csv("/content/votacao_secao-zona_2024_sp_283_ana_nice_martins_de_carvalho.csv", encoding='latin1', sep=';')
except Exception as e:
    print(f"Erro ao ler o arquivo: {e}")
    exit()

# --- Agregar votos por local de votação ---
df_votos_local = df_votos_secao.groupby('nm_local_votacao')['qt_votos'].sum().reset_index()

# --- Ordenar por quantidade de votos ---
df_ordenado = df_votos_local.sort_values('qt_votos', ascending=True)

# --- Criar cores em degradê (ajuste as cores conforme desejado) ---
cores = list(mcolors.LinearSegmentedColormap.from_list("", ["lightblue", "royalblue"])(np.linspace(0, 1, len(df_ordenado))))

# --- Plotagem do gráfico com cores em degradê ---
plt.figure(figsize=(12, 8))
barras = plt.barh(df_ordenado['nm_local_votacao'], df_ordenado['qt_votos'], color=cores)

# Adicionar os valores no gráfico
for barra in barras:
    width = barra.get_width()
    plt.text(width, barra.get_y() + barra.get_height()/2, f'{width:.0f}', va='center', ha='left', color='black', fontsize=10)

plt.xlabel('Quantidade de Votos', fontsize=12)
plt.ylabel('Local de Votação', fontsize=12)
plt.title(f'Votos de {NOME_CANDIDATA} na Zona {ZONA_ELEITORAL} ({ANO_ELEICAO})', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Adicionar o total de votos no gráfico
plt.text(0.98, 0.1, f"Total de votos: {TOTAL_VOTOS_ANA_NICE}", transform=plt.gca().transAxes, ha='right', va='top', fontsize=12, bbox=dict(facecolor='white', alpha=0.9))

plt.tight_layout()
plt.show()

# --- Criar e exibir a tabela ---
print("\nTabela de Votos por Escola:")
df_ordenado['Votos'] = df_ordenado['qt_votos']
df_tabela = df_ordenado[['nm_local_votacao', 'Votos']]
print(df_tabela)
print(f"\nTotal de votos: {df_tabela['Votos'].sum()}")
```
