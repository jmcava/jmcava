# **Zona 284 - eleição 2022 - deputado estadual 50 mais votados**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o arquivo CSV com ajustes de codificação, delimitador e escolha de colunas
df = pd.read_csv('/content/votacao_candidato.csv', encoding='latin-1', delimiter=';')

# Filtrar as colunas necessárias
df_filtrado = df[['Ano de eleição', 'Cargo', 'Município', 'Zona', 'Nome candidato', 'Votos nominais']]

# Aplicar o filtro para os dados de 2022, Deputado Estadual, Município "SÃO BERNARDO DO CAMPO" e Zona 284
df_filtrado = df_filtrado[
    (df_filtrado['Ano de eleição'] == 2022) &
    (df_filtrado['Cargo'] == 'Deputado Estadual') &
    (df_filtrado['Município'] == 'SÃO BERNARDO DO CAMPO') &
    (df_filtrado['Zona'] == 284)
]

# Ordenar os dados pelos votos nominais, pegando os 50 candidatos mais votados
df_filtrado = df_filtrado.sort_values(by='Votos nominais', ascending=False).head(50)

# Dividir os dados em dois blocos de 25 candidatos
df_bloco_1 = df_filtrado.iloc[:25]
df_bloco_2 = df_filtrado.iloc[25:]

# Criar o gráfico para o Bloco 1 (primeiros 25 candidatos)
plt.figure(figsize=(10, 8))
plt.barh(df_bloco_1['Nome candidato'], df_bloco_1['Votos nominais'], color='skyblue')

# Adicionar os valores de votos ao lado de cada barra no Bloco 1
for index, value in enumerate(df_bloco_1['Votos nominais']):
    plt.text(value, index, f' {value}', va='center', fontsize=10)

plt.title('Candidatos a Deputado Estadual Mais Votados - Bloco 1 - São Bernardo do Campo (Zona 284) - 2022', fontsize=14)
plt.xlabel('Votos Nominais', fontsize=12)
plt.ylabel('Candidatos', fontsize=12)
plt.gca().invert_yaxis()  # Inverter a ordem das barras (para o maior valor no topo)
plt.show()

# Criar o gráfico para o Bloco 2 (últimos 25 candidatos)
plt.figure(figsize=(10, 8))
plt.barh(df_bloco_2['Nome candidato'], df_bloco_2['Votos nominais'], color='lightcoral')

# Adicionar os valores de votos ao lado de cada barra no Bloco 2
for index, value in enumerate(df_bloco_2['Votos nominais']):
    plt.text(value, index, f' {value}', va='center', fontsize=10)

plt.title('Candidatos a Deputado Estadual Mais Votados - Bloco 2 - São Bernardo do Campo (Zona 284) - 2022', fontsize=14)
plt.xlabel('Votos Nominais', fontsize=12)
plt.ylabel('Candidatos', fontsize=12)
plt.gca().invert_yaxis()  # Inverter a ordem das barras (para o maior valor no topo)
plt.show()
```

```python
def gerar_graficos(zona, arquivo_csv):
    import pandas as pd
    import matplotlib.pyplot as plt

    # Carregar o arquivo CSV
    df = pd.read_csv(arquivo_csv, encoding='latin-1', delimiter=';')

    # Filtrar as colunas necessárias
    df_filtrado = df[['Ano de eleição', 'Cargo', 'Município', 'Zona', 'Nome candidato', 'Votos nominais']]

    # Aplicar os filtros necessários
    df_filtrado = df_filtrado[
        (df_filtrado['Ano de eleição'] == 2022) &
        (df_filtrado['Cargo'] == 'Deputado Estadual') &
        (df_filtrado['Município'] == 'SÃO BERNARDO DO CAMPO') &
        (df_filtrado['Zona'] == zona)
    ]

    # Ordenar pelos votos nominais e dividir em dois blocos
    df_filtrado = df_filtrado.sort_values(by='Votos nominais', ascending=False).head(50)
    df_bloco_1 = df_filtrado.iloc[:25]
    df_bloco_2 = df_filtrado.iloc[25:]

    # Gráfico Bloco 1
    plt.figure(figsize=(10, 8))
    plt.barh(df_bloco_1['Nome candidato'], df_bloco_1['Votos nominais'], color='skyblue')
    for index, value in enumerate(df_bloco_1['Votos nominais']):
        plt.text(value, index, f'{value}', va='center', ha='left', fontsize=9)
    plt.title(f'Candidatos a Deputado Estadual Mais Votados - Bloco 1 - Zona {zona} - 2022')
    plt.xlabel('Votos Nominais')
    plt.ylabel('Candidatos')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig(f'/content/bloco_1_deputado_estadual_{zona}_2022.png')
    plt.show()
    plt.close()

    # Gráfico Bloco 2
    plt.figure(figsize=(10, 8))
    plt.barh(df_bloco_2['Nome candidato'], df_bloco_2['Votos nominais'], color='lightcoral')
    for index, value in enumerate(df_bloco_2['Votos nominais']):
        plt.text(value, index, f'{value}', va='center', ha='left', fontsize=9)
    plt.title(f'Candidatos a Deputado Estadual Mais Votados - Bloco 2 - Zona {zona} - 2022')
    plt.xlabel('Votos Nominais')
    plt.ylabel('Candidatos')
    plt.gca().invert_yaxis()
    plt.tight_layout()
    plt.savefig(f'/content/bloco_2_deputado_estadual_{zona}_2022.png')
    plt.show()
    plt.close()

# Gerar os gráficos para a Zona 174
gerar_graficos(174, '/content/votacao_candidato_174.csv')
```

```python
import pandas as pd

# Caminho do arquivo CSV
file_path = '/content/votacao_candidato.csv'

# Carregar o arquivo com a codificação adequada e o separador correto
try:
    df = pd.read_csv(file_path, encoding='utf-8', sep=';')
except UnicodeDecodeError:
    df = pd.read_csv(file_path, encoding='latin1', sep=';')

# Exibir as colunas para confirmação
print("Colunas disponíveis no arquivo:")
print(df.columns)

# Filtrar apenas os candidatos eleitos e reeleitos com base na situação de totalização
# Certifique-se de ajustar os valores de acordo com a nomenclatura do seu arquivo
df_eleitos = df[df['Situação totalização'].str.contains('Eleito', case=False, na=False)]

# Selecionar colunas relevantes
df_eleitos = df_eleitos[['Nome candidato', 'Partido', 'Cargo', 'Votos nominais']]

# Ordenar por votos nominais de forma decrescente
df_eleitos_sorted = df_eleitos.sort_values(by='Votos nominais', ascending=False)

# Mostrar os primeiros registros para validação
print(df_eleitos_sorted.head())

# Salvar o resultado em um novo arquivo CSV
df_eleitos_sorted.to_csv('/content/candidatos_eleitos.csv', index=False, encoding='utf-8')
```
