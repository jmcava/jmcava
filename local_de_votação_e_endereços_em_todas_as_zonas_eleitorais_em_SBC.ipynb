```python
import pandas as pd

caminho_arquivo = '/content/votacao_secao-zona_2024_sp_174.csv'

# Define `df` before the try block to ensure it's in scope
df = None

try:
    df = pd.read_csv(caminho_arquivo, encoding='latin-1', sep=';')  # Define df inside try block
    print(df.columns)

except FileNotFoundError:
    print(f"Erro: Arquivo não encontrado em '{caminho_arquivo}'. Verifique o caminho.")
except pd.errors.ParserError:
    print(f"Erro: Problema ao analisar o arquivo CSV. Verifique se o formato e os separadores estão corretos.")
except UnicodeDecodeError:
    print(f"Erro: Problema de codificação. Tente especificar a codificação correta (e.g., encoding='latin1', encoding='iso-8859-1').")
except Exception as e:
    print(f"Erro inesperado: {e}")

# Check if df was successfully created before printing columns
if df is not None:
    print(df.columns)
else:
    print("DataFrame 'df' could not be created due to previous errors.")
```

```python
import pandas as pd

# ... (seu código para ler o CSV com a codificação correta: 'latin-1' no caso anterior)

zonas_desejadas = [174, 283, 284, 296, 409, 414]


escolas_por_zona = {}

for index, row in df.iterrows():
    zona = row['nr_zona']
    escola = row['nm_local_votacao']
    endereco = row['ds_local_votacao_endereco']

    if zona in zonas_desejadas:  # Filtra pelas zonas desejadas
        if zona not in escolas_por_zona:
            escolas_por_zona[zona] = {}  # Dicionário para armazenar escola e endereço

        escolas_por_zona[zona][escola] = endereco # Guarda o endereço junto com o nome da escola




# Imprime o relatório
for zona, escolas in escolas_por_zona.items():
    print(f"\nZona Eleitoral: {zona}")
    for escola, endereco in escolas.items():
        print(f"  - {escola}: {endereco}")
```
