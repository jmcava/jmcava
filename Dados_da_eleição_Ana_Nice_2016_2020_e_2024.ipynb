## Votos recebidos nas escolas do Rudge Ramos zona 284°

```python
import pandas as pd

# Dados da votação da Ana Nice em 2020
data_2020 = {
    'Escola': ['EXTERNATO RIO BRANCO', 'EMEBE, NEUSA BASSETTO', 'EE, LAURO GOMES DE ALMEIDA',
               'EMEB, PROF, KAZUÊ FUZINÁKA', 'EMEB, PROF, OTÍLIO DE OLIVEIRA', 'EE, PROF, AMADEU OLIVÉRIO',
               'EMEB, VIRIATO CORREIA', 'EE, PROFA, CYNIRA PIRES SANTOS'],
    'Votos_2020': [29, 33, 31, 26, 49, 52, 33, 63]
}

# Total de votos em 2016 (de acordo com sua especificação)
total_votos_2016 = 284

# Calculando a diferença de votos e a porcentagem de crescimento
diferenca_votos = [0] * len(data_2020['Escola'])
porcentagem_crescimento = [0] * len(data_2020['Escola'])

for i, escola in enumerate(data_2020['Escola']):
    # Defina os votos de 2016 para cada escola
    if escola == 'EXTERNATO RIO BRANCO':
        votos_2016 = 18
    elif escola == 'EMEBE, NEUSA BASSETTO':
        votos_2016 = 15
    elif escola == 'EE, LAURO GOMES DE ALMEIDA':
        votos_2016 = 27
    elif escola == 'EMEB, PROF, KAZUÊ FUZINÁKA':
        votos_2016 = 14
    elif escola == 'EMEB, PROF, OTÍLIO DE OLIVEIRA':
        votos_2016 = 31
    elif escola == 'EE, PROF, AMADEU OLIVÉRIO':
        votos_2016 = 16
    elif escola == 'EMEB, VIRIATO CORREIA':
        votos_2016 = 13
    elif escola == 'EE, PROFA, CYNIRA PIRES SANTOS':
        votos_2016 = 25

    # Calculando a diferença de votos e a porcentagem de crescimento
    diferenca_votos[i] = data_2020['Votos_2020'][i] - votos_2016
    porcentagem_crescimento[i] = (diferenca_votos[i] / total_votos_2016) * 100

# Criando DataFrame
df = pd.DataFrame(data_2020)

# Adicionando colunas com os votos de 2016, a diferença de votos e a porcentagem de crescimento
df['Votos_2016'] = [18, 15, 27, 14, 31, 16, 13, 25]
df['Diferenca_Votos'] = diferenca_votos
df['Porcentagem_Crescimento'] = porcentagem_crescimento

# Exibindo a tabela
print(df)
```

Para calcular a probabilidade de crescimento ou manutenção dos votos da Ana Nice nas eleições de 2024, podemos usar a média das porcentagens de crescimento obtidas nas duas eleições anteriores. Vamos calcular a média das porcentagens de crescimento:

(3.873239 + 6.338028 + 1.408451 + 4.225352 + 6.338028 + 12.676056 + 7.042254) / 7 = 6.029809%

# A média das porcentagens de crescimento é aproximadamente 6.03%.

```python
import pandas as pd

# Definindo os votos recebidos em 2020
votos_2020 = {
    "EXTERNATO RIO BRANCO": 29,
    "EMEBE, NEUSA BASSETTO": 33,
    "EE, LAURO GOMES DE ALMEIDA": 31,
    "EMEB, PROF, KAZUÊ FUZINÁKA": 26,
    "EMEB, PROF, OTÍLIO DE OLIVEIRA": 49,
    "EE, PROF, AMADEU OLIVÉRIO": 52,
    "EMEB, VIRIATO CORREIA": 33,
    "EE, PROFA, CYNIRA PIRES SANTOS": 63
}

# Calculando a média das porcentagens de crescimento
porcentagens_crescimento = [3.873239, 6.338028, 1.408451, 4.225352, 6.338028, 12.676056, 7.042254]
media_crescimento = sum(porcentagens_crescimento) / len(porcentagens_crescimento)

# Calculando os votos previstos para 2024
votos_previstos_2024 = {}
for escola, votos in votos_2020.items():
    votos_previstos = votos * (1 + media_crescimento / 100)
    votos_previstos_arredondados = round(votos_previstos)
    votos_previstos_2024[escola] = votos_previstos_arredondados

# Criando o DataFrame com os votos previstos para 2024
df_votos_previstos = pd.DataFrame(list(votos_previstos_2024.items()), columns=['Escola', 'Votos_Previstos_2024'])

# Calculando o total de votos previstos para 2024
total_votos_previstos_2024 = sum(votos_previstos_2024.values())

# Adicionando a linha com o total ao DataFrame
total_row = pd.DataFrame([['Total', total_votos_previstos_2024]], columns=['Escola', 'Votos_Previstos_2024'])
df_votos_previstos = pd.concat([df_votos_previstos, total_row], ignore_index=True)

# Renomeando a coluna para 'Meta_Total_2024'
df_votos_previstos.rename(columns={'Votos_Previstos_2024': 'Meta_Total_2024'}, inplace=True)

# Exibindo o DataFrame resultante
print(df_votos_previstos)
```

```python
import pandas as pd

# Dados fornecidos
data = {
    "Escola": ["EXTERNATO RIO BRANCO", "EMEBE, NEUSA BASSETTO", "EE, LAURO GOMES DE ALMEIDA",
               "EMEB, PROF, KAZUÊ FUZINÁKA", "EMEB, PROF, OTÍLIO DE OLIVEIRA", "EE, PROF, AMADEU OLIVÉRIO",
               "EMEB, VIRIATO CORREIA", "EE, PROFA, CYNIRA PIRES SANTOS"],
    "Votos_2020": [29, 33, 31, 26, 49, 52, 33, 63],
    "Votos_2016": [18, 15, 27, 14, 31, 16, 13, 25],
    "Diferenca_Votos": [11, 18, 4, 12, 18, 36, 20, 38]
}

# Criando o DataFrame
df = pd.DataFrame(data)

# Calculando a média das porcentagens de crescimento
porcentagens_crescimento = [3.873239, 6.338028, 1.408451, 4.225352, 6.338028, 12.676056, 7.042254]
media_crescimento = sum(porcentagens_crescimento) / len(porcentagens_crescimento)

# Calculando os votos previstos para 2024
df['Meta_Total_2024'] = df['Votos_2020'] * (1 + media_crescimento / 100)
df['Meta_Total_2024'] = df['Meta_Total_2024'].round()

# Calculando os totais
total_votos_2020 = df['Votos_2020'].sum()
total_votos_2016 = df['Votos_2016'].sum()
total_meta_total_2024 = df['Meta_Total_2024'].sum()

# Adicionando a linha com os totais ao DataFrame
total_row = pd.DataFrame([['Total', total_votos_2020, total_votos_2016, total_meta_total_2024]],
                         columns=['Escola', 'Votos_2020', 'Votos_2016', 'Meta_Total_2024'])
df = pd.concat([df, total_row], ignore_index=True)

# Descartando a coluna Diferenca_Votos
df = df.drop(columns=['Diferenca_Votos'])

# Reordenando as colunas
df = df[['Escola', 'Votos_2016', 'Votos_2020', 'Meta_Total_2024']]

# Exibindo o DataFrame resultante
print(df)
```
